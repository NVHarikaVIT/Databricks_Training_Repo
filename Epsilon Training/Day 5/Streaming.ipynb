Cons of batch processing

```python
df=spark.read.csv("/Volumes/dev/naval/raw/stream/",header=True,inferSchema=True)
df.write.mode("overwrite").saveAsTable("dev.naval.streaming_b") # if millions of files are streamed, same number of truncate and rewrite happens
```

```python
df=spark.read.csv("/Volumes/dev/naval/raw/stream/",header=True,inferSchema=True)
df.write.mode("append").saveAsTable("dev.naval.streaming_b") # if millions of files are streamed, records from previous files are appened same number of times.
```

### Stream Processing

```python
# defining schema
schema="Id integer, Name string, Gender string, Salary integer,Country string, Date string"
```

```python
df = spark.readStream.schema(schema).csv("/Volumes/dev/naval/raw/stream/",header=True)
```

```python
display(df)
```

```python
# Naval has added one more file.
df.writeStream.option("checkpointLocation","/Volumes/dev/harika/raw_data/checkpoint").trigger(availableNow=True).toTable("dev.harika.streaming_data")
```

```python
%sql
select * from dev.harika.streaming_data;
```

-- The major drawback of structured streaming is, when we ingest new data which follows different schema than the predefined, it won't add consistent values and won't add new columns if there are more.

-- At that time, it is better to use the concept of AutoLoader which provides flexible schema option (but it is complex.)

-- So, we can use Lakeflow Pipelines.
