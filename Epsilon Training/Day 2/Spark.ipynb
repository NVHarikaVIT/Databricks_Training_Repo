Spark is an open-source parallel processing engine

Datasets
1. RDD
2. Dataframe

Spark is a Lazy Evaluator. It doesn't show performance metrics when we do Transformations (like read, write).
It only shows performance metrics when we display the output (aka. when we perform actions).

Prior to performing actions, Spark will plan the execution. When we call any action, then Spark actually executes.

```python
input_path="/Volumes/dev/naval/raw/products.csv"
catalog="dev"
schema="harika"
tablname="products"
```

```python
df=spark.read.csv(input_path,header=True,inferSchema=True)
df.write.mode('overwrite').saveAsTable(f"{catalog}.{schema}.{tablname}")
```

```python
df=spark.read.table("dev.harika.products")
df.display()
```

```python
#ETL process
from pyspark.sql.functions import *

# reading data
data = spark.read.csv(input_path,header=True,inferSchema=True)

# transformation -- updating the table with metadata information
new_data = data.withColumn("ingestion_date", current_timestamp()).withColumn("filePath", col("_metadata.file_path"))

# load the transformed data
new_data.write.saveAsTable(f"{catalog}.{schema}.{tablname}_metadata")
```

```python
# reading the new data
df=spark.read.table("dev.harika.products_metadata")
df.display()
```

```python
# ETL with JSON file.

# read data
df = spark.read.json("/Volumes/dev/naval/raw/drivers.json")

# transformation
df1 = (df.withColumn("forename", col("name.forename")).withColumn("surname", col("name.surname")).withColumn("ingestion_date", current_timestamp()).withColumn("filePath", col("_metadata.file_path")).drop("name"))

# load data
df1.write.mode("overwrite").saveAsTable("dev.harika.drivers_metadata")
```

```python
df = spark.read.table("dev.harika.drivers_metadata")
df.display()
```
