We can create catalog via UI or syntax.

CREATE CATALOG IF NOT EXISTS dev;

Create schema via 

CREATE SCHEMA if not exists dev.harika 

or

USE dev; 

CREATE SCHEMA if not exists harika

```sql
-- Creating a schema
CREATE SCHEMA IF NOT EXISTS dev.harika;
```

```sql
-- Create a table
USE dev.harika;
CREATE TABLE IF NOT EXISTS demo_table(id INT, name STRING, birth_date DATE);
```

```sql
-- Insert a few records
INSERT INTO demo_table VALUES (1, 'John', '2004-03-10'), (2, 'Mary', '2002-12-19'), (3,'Shyam','2001-01-01');
-- Display the data
SELECT * from demo_table;
```

Batch ingestion of data via CREATE TABLE AS SELECT (CTAS) Technique

```sql
CREATE TABLE IF NOT EXISTS dev.harika.housing_data AS
select * FROM read_files('/Volumes/dev/naval/raw/housing.csv', format=>'csv');
```

```sql
select * from dev.harika.housing_data;
```
